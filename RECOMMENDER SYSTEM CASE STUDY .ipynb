```python
# first of all merge both table for doing any analysis.

jointdata = ratings.merge( movies, on = 'movieId' , how = 'left' )
```

```python
jointdata.columns
```

```python
jointdata.title.unique()
```

```python
jointdata.head(10)
```

```python
# FINDING AVERAGE RATINGS PER USERID
average_ratings = pd.DataFrame(jointdata.groupby('title')['rating'].mean())
```

```python
average_ratings.head(10) # so we can see avg ratings for movie based on movie title.
```

```python
# now will find out total number of ratings for each movie.
```

```python
average_ratings['Total_Ratings'] = pd.DataFrame(jointdata.groupby('title')['rating'].count())
```

```python
average_ratings.head()
```

```python
# now calculate the coorelation by creating new variable.The value of the matrix represent the ratings by each user for each movie.

movie_user = jointdata.pivot_table(index = 'userId', columns = 'title', values = 'rating')
```

```python
movie_user.head()
```

```python
movie_user.tail()
```

```python
# now will select any movie title for checking "Recommender System".
# for finding coorelation of one movies with another movie we will pass all the ratings of the one movie to the coor with method 
#  of the pandas dataframe.
# pandas dataframe method compute the pairwise coorelation between rows and column of a dataframe with rows of columns of series 
# or dataframe.
```

```python
correlations = movie_user.corrwith(movie_user['Grumpier Old Men (1995)'])
```

```python
correlations
```

```python
correlations1 = movie_user.corrwith(movie_user['Toy Story (1995)'])
```

```python
correlations1
```

```python
# now after watching correlations we will remove all empty value and merge the total ratings to the corellation table.
```

```python
recommendation = pd.DataFrame(correlations1, columns = ['Correlation'])
```

```python
recommendation
```

```python
recommendation.dropna(inplace = True)
```

```python
recommendation
```

```python
recommendation = recommendation.join(average_ratings['Total_Ratings'])
```

```python
recommendation
```

```python
#now will test the recommender system
#now filter out all the movies with correlation value to toy story and with at least 100 ratings.
```

```python
recc = recommendation[recommendation['Total_Ratings']>100].sort_values('Correlation', ascending = False ).reset_index()
```

```python
recc
```

```python
# now merge the movie dataset to verifying the recommendations.
```

```python
recc = recc.merge(movies, on = 'title', how = 'left')
```

```python
recc
```

```python
# we can see the top movie like incredible, finding nemo , alladin having good correlation with toy story.
```

```python
# now same we can find recommendation like this for any other movie too .
# now i am going to match correlation of grumpier old man with other movie to see which other movie having same genres to watch.
```

```python
correlations
```

```python
#now again will remove empty value and merge the total ratings to the correlation table.
```

```python
recommendation_for_GOM = pd.DataFrame(correlations , columns = ['Correlation'])
```

```python
recommendation_for_GOM.dropna(inplace = True)
```

```python
recommendation_for_GOM = recommendation_for_GOM.join(average_ratings['Total_Ratings'])
```

```python
recommendation_for_GOM.head(2)
```

```python
# testing the recommender system and filter all the movies with correlation with grumpier old man with atleast 100 ratings.
```

```python
recc1 = recommendation_for_GOM[recommendation_for_GOM['Total_Ratings']>100].sort_values('Correlation', ascending = False).reset_index()
```

```python
recc1.head(5)
```

```python
#now merge movie table too for verifying the recommendations.
```

```python
recc1 = recc1.merge(movies , on = 'title', how = 'left')
```

```python
recc1.head(15)
```
